### **Data preprocessing (making the dataframes smaller)**

```python
# NOTE: you cannot just remove the rows containing the null values from this dataset because virtually
# almost every row has a missing value which makes the shape of the dataframe after removing
# the rows to just 46 rows from the original 13 million. The next best approach is to just
# fill the missing data in with things like the mean or mode of the column.
```

```python
# First get the column names that train_df and test_df both share so I can convert that to English
shared_cols=train_df.columns.intersection(test_df.columns)
print("Shared column names:", list(shared_cols))
```

```python
# Rename columns to english so I can understand it better for both dataframes

# Use the column names that they both share which would be 
english_colnames={"fecha_dato":"calendar_date", "ncodpers":"cust_id", "ind_empleado":"emp_index", "pais_residencia":"residence", "sexo":"sex",
                 "age":"age", "fecha_alta":"first_date", "ind_nuevo":"new_cust", "antiguedad":"seniority", "indrel":"is_primary", "ult_fec_cli_1t":"last_primary_date",
                 "indrel_1mes":"cust_type", "tiprel_1mes":"cust_rel_type", "indresi":"residence_index", "indext":"foreigner_index", "conyuemp":"spouse_index",
                 "canal_entrada":"channel", "indfall":"deceased_index", "tipodom":"address_type", "cod_prov":"province", "nomprov":"province_name", "ind_actividad_cliente":"active_index",
                 "renta":"income","segmento":"segment"}

train_df.rename(english_colnames, axis=1, inplace=True)
test_df.rename(english_colnames, axis=1, inplace=True)

# Check to see if it renamed
test_df.columns
```

```python
# Check to see datatypes again because I'm too lazy to scroll up

print("Train_df datatypes:\n")
print(train_df.dtypes)
print("Test_df datatypes:\n")
print(test_df.dtypes)
```

```python
# Convert the columns to respective datatypes to make sure the numbers and dates are not strings
# Converts dates to dates and objects to appropriate integers/floats

train_df.calendar_date=pd.to_datetime(train_df.calendar_date, errors="coerce")
train_df.first_date=pd.to_datetime(train_df.first_date, errors="coerce")
train_df.income=pd.to_numeric(train_df.income, errors="coerce")

test_df.calendar_date=pd.to_datetime(test_df.calendar_date, errors="coerce")
test_df.first_date=pd.to_datetime(test_df.first_date, errors="coerce")
test_df.income=pd.to_numeric(test_df.income, errors="coerce")
```

```python
# Check how many nulls there are because I'm too lazy to scroll up again
train_df.isnull().sum()/train_df.shape[0]*100
```

```python
# Remove spouse_index and last_primary_date because there are too many null variables
train_df.drop(["last_primary_date","spouse_index"], axis=1, inplace=True)
test_df.drop(["last_primary_date","spouse_index"], axis=1, inplace=True)

# Replace the null values with the mean/median/mode
## For the objects/strings/features in these datasets, I will do the mode which takes the most frequent value because the goal
## of this challenge is to get the most popular output anyways. And then for the numerical stuff, I will fill
## it with the average.

categories=["emp_index", "residence", "sex", "first_date", "new_cust", "is_primary", "cust_type", "cust_rel_type", "province", 
           "province_name", "active_index", "channel", "segment"]

for features in categories:
    train_df.loc[train_df[features].isnull(), features]=train_df[features].value_counts().index[0]
    test_df.loc[test_df[features].isnull(), features]=test_df[features].value_counts().index[0]

# test_df[features].value_counts().index[0] represents the MODE

# Then for income, replace with the average income (except it is a wide range so we can group by the area
# they came from (province code) and if they graduated college or not)

train_df["income"]=train_df.groupby("segment")["income"].transform(lambda x:x.fillna(x.median()))
test_df["income"]=test_df.groupby("segment")["income"].transform(lambda x:x.fillna(x.median()))
train_df["income"]=train_df.groupby("province")["income"].transform(lambda x:x.fillna(x.median()))
test_df["income"]=test_df.groupby("province")["income"].transform(lambda x:x.fillna(x.median()))
```
